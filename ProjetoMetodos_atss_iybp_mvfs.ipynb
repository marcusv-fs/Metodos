# Modelo
Encontrar um bom modelo matemático que descreve eventos na natureza não é uma tarefa fácil. Entretanto, apesar de não serem completamente precisos, algumas informações que os modelos aproximados nos sugerem são de grande importância para avaliar uma potencial ameaça. 

Pensando nisso, em 1927, desenvolvido por Kermarck e Mckendrick surgiu o modelo SIR para avaliação de modelos pandêmicos. Nele, são avaliadas doenças onde o indivíduo adquire imunidade após curar-se da doença. Para isso, são utilizados três parâmetros, as pessoas suscetíveis (S) à infecção, a quantidade de pessoas infectadas (I) e as pessoas que já se recuperaram (R) e, portanto, não podem mais se infectar nem transmitir. 

Contudo, esse não é um modelo tão preciso, visto que em muitas doenças uma pessoa pode se reinfectar, além de não considerar fatores como exposição, a quantidade de pessoas mortas pela doença e outros fatores que o aproximam mais do que acontece no mundo real.

Nesse projeto é utilizado um modelo com mais precisão, o SEIRD, no qual são considerados parâmetros extras para tentar aproximar o modelo matemático das curvas reais de desenvolvimento da doença. Esse modelo já foi utilizado em outros estudos, os quais serviram de inspiração e é, basicamente, uma evolução do SIR. Para isso, são usados os seguintes parâmetros:

\

•	S(t): quantidade de pessoas suscetíveis a contrair o vírus.

•	E(t): a quantidade de pessoas expostas.

•	I(t): representa a taxa média de pessoas infectadas.

•	R(t): a quantidade de pessoas recuperadas.

•	D(t): a quantidade de pessoas mortas.


Os parâmetros que serão utilizados:

•	t: Tempo

•	N: Tamanho da população

•	${\gamma}$: é a taxa média de recuperação da doença.

•	β: representa a taxa de transmissão da doença; Onde β(t) * n nos resulta um indivíduo infectado entrando em contato com n pessoas.

•	${\alpha}$: A taxa de letalidade é calculada dividindo-se o número de mortes de uma doença específica durante um período de tempo definido pelo número de indivíduos diagnosticados com a doença durante esse período;

•	σ: taxa média do período latente;

\


Essas serão as fórmulas utilizadas nesse estudo:

| ${\displaystyle \frac{dS}{dt} }$= ${ -\frac{\beta \cdot S \cdot I}{N} }$

| ${\displaystyle \frac{dE}{dt} }$ = ${ \frac{\beta \cdot S \cdot I}{N}  - \sigma \cdot E}$

| ${\displaystyle \frac{dI}{dt} }$ = ${ σ \cdot E - \gamma \cdot I- α\cdot I}$


| ${\displaystyle \frac{dR}{dt} }$ =  ${ \gamma \cdot I}$

| ${\displaystyle \frac{dD}{dt} }$ =  ${ α  \cdot I}$

\

Para levar em consideração a população dos vacinados, é preciso incluir uma variação do modelo SIR, o SIRV + E. Fazendo uma comparação com o modelo SEIRD utilizado anteriormente, nele é considerado o fator "D", onde se considera o número de mortos pela doença, no entanto, agora, quando o indivíduo for vacinado, não há mais mortes. Portanto, as novas equações serão:

| N = S(t) + I(t) + R(t) + V(t)

| ${\displaystyle \frac{dS(t)}{dt}}$ = ${  Ωr\cdot R(t) - λ\cdot S(t) + Ωv\cdot V(t) -  Φ \cdot S(t)}$ 

| ${\displaystyle \frac{dV(t)}{dt}}$ = ${   Φ  \cdot S - Ωv\cdot V(t) -(1 - ψ)⋅ λ\cdot V(t)}$ 

| ${\displaystyle \frac{dI(t)}{dt}}$ = ${  λ\cdot S(t) + (1-ψ)\cdot λ \cdot V(t) - \gamma \cdot I(t)}$ 

| ${\displaystyle \frac{dR(t)}{dt}}$ = ${  \gamma\cdot I(t) - Ωr\cdot R(t)}$ 

Onde: 

> N: é a população total

> S(t): representa suscetível a ter a doença no tempo t

> I(t): representa a população infectada com o vírus no tempo t

> R(t): representa a população removida no tempo t, dizer que uma população foi removida nesse modelo significa que ou ficou recuperada da doença.

> V(t): representa a população vacinada no tempo t

> λ: representa a taxa de infecção por pessoa e é dada por λ = ${\frac{BI}{N}}$

> ψ: representa a taxa de vacinação e eficácia das vacinas, ψ = eficácia media * Φ

> Φ: taxa de vacinação

> Ωr: representa a taxa com a qual os indivíduos recuperados voltam para o grupo dos suscetíveis devido à perda de imunidade.

> Ωv: representa a taxa com a qual os indivíduos vacinados voltam para o grupo dos suscetíveis devido à perda de imunidade.

Esse modelo representa a dinâmica
de propagação do vírus levando em conta a campanha
de imunização da população e, portanto, pode-se com o
esforço da vacinação, planejar melhores estratégias para
controlar a dinâmica de propagação do vírus.

#Modelo SEIRD

```python
def derivadas(y, t, beta, gama, sigma, alfa):
  
  S, E, I, R, D = y 
  dSdt = -beta * S *I
  dEdt = beta*S*I -sigma*E
  dIdt = sigma*E -gama*I- alfa*I
  dRdt = gama*I
  dDdt = alfa*I

  return dSdt, dEdt, dIdt, dRdt, dDdt
```

```python
def plot_result(seird_result, t):
  y1, y2, y3, y4, y5 = seird_result.T
  figure = plt.figure(figsize=(40, 10))
  ax = figure.add_subplot(121)
  ax.plot(t, y1, 'b', alpha=0.5, lw=2, label='Suscetíveis')
  ax.plot(t, y2, 'r', alpha=0.5, lw=2, label='Expostos')
  ax.plot(t, y3, 'g', alpha=0.5, lw=2, label='Infectados')
  ax.plot(t, y4, 'y', alpha=0.5, lw=2, label='Recuperados')
  ax.plot(t, y5, 'm', alpha=0.5, lw=2, label='Mortos')
  ax.set_xlabel('dias')
  ax.set_ylabel('Pessoas')
  ax.set_title('Covid em pernambuco')
  legend = ax.legend();
  legend.get_frame().set_alpha(0.5)
  figure.show
```

#Modelo SIRV

```python
def derivada_SIRV(y0, t, Lambda, gamma, psi, phi, wV, wR):
  S, I, R, V = y0
  
  dStdt = wR*R - Lambda*S + wV*V - phi*S
  dItdt = (psi-1) * Lambda * V * I - wV*V + phi*S
  dRtdt = gamma*I - wR * R
  dVtdt = (psi-1) * Lambda*V*I - wV* V + phi*S

  return dStdt, dItdt, dRtdt, dVtdt
```

> Jassen ⇒ foram distribuídas 305.510 doses que podem ser aplicadas em 1.73% da população. Esta vacina possui uma eficácia de 66.9%;

> Pfizer ⇒ foram distribuídas 7.743.720 doses que podem ser aplicadas em 43.89% da população. Esta vacina possui uma eficácia de 91.30%.

Dessa forma, fazendo a média ponderada de cada vacina, podemos obter, a média total ${\alpha}$ = 74,24% = 0,7424. Esta média será utilizada no da eficação total da vacina, denotada anteriormente por: "ψ".

```python
N = 9676793 #[1]populacao de PE em 2021 segundo IBGE
I = 552289 #[2]quantidade de pessoas infectadas em 30/06/2021
R = 467422 #[3]numero de recuperados em 30/06/2021
V = 7447043 #[8]numero de vacinados

alpha = 0.7424 #eficacia media
beta = (1/14)/N

Lambda = (beta*I)/N # λ: representa a taxa de infecção 
gamma = 1/14        # Taxa de recuperação
phi = 0.008466     #  'Φ' taxa de vacinação
psi = phi *  alpha     # 'ψ' eficácia da vacina
omegaV = 0.001170 # 'Ω'Imunidade da vacina cai 50% ao ano
omegaR = 0.001897 # 'Ω'Imunidade de recuperação cai 10% a cada 90 dias

S = N -I -R -V
y0 = S, I, R, V #vetor que armazena as condições iniciais do sistema

#suposicao
t = np.linspace(0, 395, 396)# 30/06/2021-30/06/2022
resSirv = odeint(derivada_SIRV, y0, t, args = (Lambda, gamma, psi, phi, omegaV, omegaR))
```

```python
plot_resultSirv(resSirv, t)
```

```python
plot_resultsSIRV(resSirv, t)
```

Houve uma "explosão" no número de recuperados, pois, mesmo as pessoas sendo infectadas, quando há um aumento na vacinação, elas se recuperam mais facilmente.

Na prática, não ocorreu bem assim. Durante a pandemia as políticas públicas de isolamento e utilização de máscaras resultaram em uma diminuição no número de contágios. Contudo, ocorreram fenômenos que se assemelham a ondas de infecção, causados pela diminuição na intensidade do isolamento pela população, geralmente em períodos de festividades ou após um período com baixas taxas de mortes. Isso fazia com que o número de casos aumentasse e, graças a isso, os governos aumentavam o grau de isolamento e as taxas mudavam, novamente, para números menores.

Outro fator importante a se salientar é que a pandemia se comportou de forma diferente nas diversas regiões do estado, isso se deve a diversos fatores, dentre eles:

  •	Político: uma esfera política defendia que a questão econômica era mais importante do que a saúde de alguns indivíduos e, por isso, os cidadãos deveriam continuar a se comportar normalmente, enquanto outros defendiam que as vidas são mais importantes e que o impacto seria ainda maior se mais pessoas morressem.

  •	Geográficos: Existem cidades com maior densidade demográfica, em que mais pessoas estão em um espaço menor, o que facilita a propagação do vírus.

  •	Econômicos: Pessoas que podem se utilizar de transportes individuais se expunham menos do que as que demandam transportes coletivos, além de poderem se dar ao luxo de trabalhar remotamente ou se afastar temporariamente dos empregos.

Esses fatores alteram bastante a disseminação do vírus e não são facilmente transpostos para um modelo matemático. Contudo, o modelo ainda é útil, mesmo não sendo tão preciso, ao informar como seria um comportamento possível da pandemia desconsiderando esses fatores mais individualistas.	


# Conclusão

A partir das análises feitas, a projeção prevista pelo modelo sem considerar vacinas e outros fatores atenuadores da doença previu uma quantidade maior de mortes e que elas ocorreriam um pico único, seguida por uma queda exponencial. Já no caso em que se estimula o isolamento e vacinação, o principal ponto a se destacar é a queda na quantidade de mortos destacada pelo aumento do número de recuperados. Entretanto, na realidade aconteceu diferente dos dois modelos propostos.

Isso se deve aos impactos das políticas de gestão da pandemia pelos governantes, como o incentivo ao isolamento e uso de máscaras, que afetam diretamente a exposição e, por tanto, o número de infectados e, consequentemente, o número de mortos. Bem como ao comportamento geral da população que mudava conforme o avanço da vacinação e queda no número de óbitos semanais.

Outro fator importante a se destacar, é que o modelo prevê o número de mortes causado diretamente pela doença, contudo, existem também as mortes causadas indiretamente e que não aparecem nos relatórios. Essas mortes foram causadas pela grande ocupação dos leitos de hospital que, pela falta de vagas, passaram a não aceitar mais pessoas, o que levou a um aumento de mortes causadas por outras doenças ou acidentes por não permitirem um tratamento. 

Além disso, outro ponto importante a se destacar é a variação do vírus. No modelo apresentado, não são considerados as diversas mutações que surgiram. Elas alteraram parâmetros como mortalidade, contágio e outros. Ademais, é válido destacar que, no mundo real, diversas dessas variações coexistiram e causaram fenômenos como infecções simultâneas por mais de uma variante, o que não são simples de serem transpostas. Por esses e outros motivos, fixamos o vírus como o que deu origem à doença.

Por fim, existem os casos em que a causa da morte era registrada como outra, apesar de ter sido provocada pela COVID para desincentivar o cuidado com a doença.

Contudo, vale ressaltar a importância desse estudo, mesmo que os resultados não demonstrem uma precisão tão grande do modelo, pois, com ele, é possível simular como seria um mundo em que a doença se espalhasse sem nenhuma restrição e ressaltar a importância das atitudes que limitam a proliferação do vírus.
